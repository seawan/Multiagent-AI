## Part I: Normal Form Games

#### Definition of a normal form game
Given sample pay off matrices,
$$
\mathbf{R}^1 = \left[\begin{matrix}
0 & 3 \\
1 &2
\end{matrix}\right]
\quad
\mathbf{R}^2 = \left[\begin{matrix}
3 & 2 \\
0 & 1
\end{matrix}\right],
$$


An $N$ player normal form game (a.k.a. Matrix Game/Stage Game) consists of:

*   A finite set of $N$ players.
*   Action set for the players: $\{\mathcal{A}_1, \mathcal{A}_2, \dots \mathcal{A}_N\}$
*   Payoff functions for the players: $u_i : \mathcal{A}_1 \times \mathcal{A}_2 \dots \times \mathcal{A}_N \to \mathbb{R}$

For example, given $N=2, \mathcal{A}_1 = \mathcal{A}_2 = \{1,2\}$, we can write the payoff function in the matrix form as follows:
$$
\mathbf{R}^1 = \left[\begin{matrix}
r^1_{11} & r^1_{12} \\
r^1_{21} & r^1_{22}
\end{matrix}\right]
\quad
\mathbf{R}^2 = \left[\begin{matrix}
r^2_{11} & r^2_{12} \\
r^2_{21} & r^2_{22}
\end{matrix}\right].
$$

If the player 1 chooses action $i$ and the player 2 chooses action $j$, then the player 1 and player2 receive the rewards $r^1_{ij}$ and $r^2_{ij}$ respectively.

Then, we can use $\pi_1 = \alpha\in [0,1] $ to represent the strategy for player 1, where $\alpha$ corresponds to the probability of player 1 selecting the first action (action 1), and $1-\alpha$ is the probability of choosing the second action (action 2). Similarly, we use $\pi_2 = \beta$ to be the strategy for player 2.

Given the pair of strategies $(\alpha, \beta)$, the possibilities of playing each action pairs are

| probabilities | column player plays 1 (with prob. $\beta$) | column player plays 2 (with prob. $1 - \beta$) |
|-|-|-|
| row player plays 1 (with prob. $\alpha$) | $\alpha\beta$ | $\alpha(1 - \beta)$ |
| row player plays 2 (with prob. $1 - \alpha$) | $(1 - \alpha) \beta$ | $(1 - \alpha)(1 - \beta)$ |


we can have the **expected payoffs** for two players. Denote $V^1(\alpha, \beta)$ and $V^2(\alpha, \beta)$ as the expected payoffs for two players respectively:

$$
\begin{align}
V^1(\alpha, \beta) &=\alpha \beta r^1_{11} & + & \alpha(1-\beta) r^1_{12} \\
    & +(1-\alpha) \beta r^1_{21} & + & (1-\alpha)(1-\beta) r^1_{22}
\end{align}      
$$  
$$
\begin{align}
V^2(\alpha, \beta) &=\alpha \beta r^2_{11} & + & \alpha(1-\beta) r^2_{12} \\
    & +(1-\alpha) \beta r^2_{21} & + & (1-\alpha)(1-\beta) r^2_{22}
\end{align}   
$$

#####Q3: Find Nash equilibria for this game through the utility plot (The Upper-envelope method).

In the two player game, A strategy pair $(\pi_{Alice}, \pi_{Bob})$ is a Nash equilibria if

1. Alice's strategy $\pi_{Alice}$ is a best response to Bob's strategy $\pi_{Bob}$
2. Bob's strategy $\pi_{Bob}$ is a best response to Alice's strategy $\pi_{Alice}$.

Recap the definition of best response: Assume that your opponent choose strategy $\pi$, and you need to decide your strategy (as a response). Then, the strategy $\pi'$ that allow you to maximize your expected payoff is your best response to $\pi$. If both the strategies are the best responses of each other, then both players will not deviate from the current strategy, which leads to an equilibria.

Let's consider the **pure strategy** first ($\alpha, \beta$ can be either 0 or 1):

For Alice:

1. If Bob choose sport movie, her best response will be sport movie (so that she can receive a reward of 3 instead of 0)
2. If Bob choose comedy, her best response will be comedy (so that she can receive a reward of 2 instead of 1)

For Bob:

1. If Alice choose sport movie, his best response will be sport movie (so that he can receive a reward of 2 instead of 1)
2. If Alice choose comedy, his best response will be comedy (so that he can receive a reward of 3 instead of 0)

Therefore, there are two pure Nash equilibria: (sport movie, sport movie) and (comedy, comedy).

Then we consider the **mixed strategy**, which is a bit more complex. Here we provide an intuitive introdution. (More rigorous introduction including the best response condition will follow in a later lab session)

Let's recap the previous result. If Bob choose sport movie, Alice can choose sport movie with an expected payoff of 3, or comedy with 0 (so she will prefer the former one). If Bob choose comedy, Alice can choose sport movie with an expected payoff of 1, or comedy with 2 (so she will prefer the latter one). Then what if Bob plays in a mixed way, with probability $\alpha$ to choose sport movie, and probability $(1 - \alpha)$ to choose comedy? Intuitively, there will be some mixed strategy $\alpha$ of Bob that make Alice "hard to choose". That is, under such a strategy, her expected payoff of choosing sport movie is equal to that of choosing comedy.

## Part II: Cournot Duopoly

The Cournot model, formulated by French economist Antoine Augustin Cournot in 1838, serves as a fundamental framework in industrial organisation and microeconomics. It examines how two firms in a **duopoly (two firms dominating
a market)**, compete by independently deciding the quantities
they produce. The single market price is influenced by the total production
from both firms. Each firm seeks to maximise its profit, assuming the other
firm's output remains unchanged. This strategic interplay results in a Nash equilibrium, where neither firm can increase its profit by altering its production unilaterally.

$$q_1^* = \arg\max_{q_1}u_1(q_1, q_2) $$

$$
-0.02q_1^* + 1.5 - 0.01q_2 = 0
